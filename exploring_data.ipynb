#### straight lines, y = ax + b

```python
x = np.arange(-5, 6, 1)
y = 2 * x + 1

f, ax = plt.subplots()
ax.plot(x, y)
```

### quadratic function:  $y = 2x^2  + 4$

```python
x = np.arange(-5, 5, 0.01)
y = 2*x*x  + 4

f, ax = plt.subplots()
ax.plot(x, y)
```
